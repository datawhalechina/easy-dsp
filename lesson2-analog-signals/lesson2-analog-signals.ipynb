```python
# 产生单个弦波Cosine Wave 1
import numpy as np
import matplotlib.pyplot as plt
# 生成时间序列
t = np.arange(0, 1, 0.001)
# 计算弦波值
y = 3 * np.cos(2 * np.pi * t + 0.5 * np.pi)

plt.plot(t, y)
plt.title('Cosine Wave')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.show()
```

```python
#第二个单个弦波Cosine Wave 2
import numpy as np
import matplotlib.pyplot as plt
# 生成时间序列
t = np.arange(0, 1, 0.001)
# 计算弦波的值
y1 = 4 * np.cos(2 * np.pi * t + 0.7 * np.pi)

plt.plot(t, y1)
plt.title('Cosine Wave')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.show()
```

```python
#两个弦波的叠加
import numpy as np
import matplotlib.pyplot as plt
# 计算两个余弦波叠加的值
y_sum = y + y1

plt.plot(t, y, label='Cosine Wave 1')
plt.plot(t, y1, label='Cosine Wave 2')
plt.plot(t, y_sum, label='Sum of Cosine Waves')
plt.title('Two Cosine Waves and Their Sum')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.legend()
plt.show()
```
