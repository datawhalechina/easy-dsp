# 周期信号的产生（以弦波为例）

```python
import numpy as np
import matplotlib.pyplot as plt

amplitude = 1.0  # 振幅
frequency = 5.0  # 频率（单位：Hz）
phase = 0.0  # 相位（单位：弧度）
duration = 1.0  # 持续时间（单位：秒）
sample_rate = 1000  # 采样率（单位：Hz）

# 生成时间序列
t = np.linspace(0, duration, int(duration * sample_rate), endpoint=False)
signal = amplitude * np.sin(2 * np.pi * frequency * t + phase)

plt.plot(t, signal)
plt.title('Signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.show()
```

# 方波信号

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal as signal

amplitude = 1.0  # 振幅
frequency = 5.0  # 频率（单位：Hz）
duty_cycle = 0.5  # 占空比（取值范围0到1）
duration = 0.5  # 持续时间（单位：秒）
sample_rate = 1000  # 采样率（单位：Hz）

# 生成时间序列
t = np.linspace(0, duration, int(duration * sample_rate), endpoint=False)

# 生成方波信号
square_wave=signal.square(2 * np.pi * 5 * t)

plt.plot(t, square_wave)
plt.title('Square Wave')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.show()
```

# 三角波

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal

amplitude = 1.0  # 振幅
frequency = 5.0  # 频率（单位：Hz）
duration = 1.0  # 持续时间（单位：秒）
sample_rate = 1000  # 采样率（单位：Hz）

# 生成时间序列
t = np.linspace(0, duration, int(duration * sample_rate), endpoint=False)

# 生成三角波信号
triangle_wave = amplitude * signal.sawtooth(2 * np.pi * frequency * t, width=0.5)

plt.plot(t, triangle_wave)
plt.title('Triangle Wave')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.show()
```

# 谐波

```python
import numpy as np
import matplotlib.pyplot as plt

# 设置第一个弦波的参数
amplitude_1 = 1.0
frequency_1 = 2.0
phase_1 = 0.0
duration = 1.0
sample_rate = 1000

# 设置第二个弦波的参数
amplitude_2 = 0.5
frequency_2 = 5.0
phase_2 = np.pi / 2

# 生成时间序列
t = np.linspace(0, duration, int(duration * sample_rate), endpoint=False)

# 生成第一个弦波信号
wave_1 = amplitude_1 * np.cos(2 * np.pi * frequency_1 * t + phase_1)

# 生成第二个弦波信号
wave_2 = amplitude_2 * np.cos(2 * np.pi * frequency_2 * t + phase_2)

# 将两个弦波相加得到谐波
harmonic_wave = wave_1 + wave_2

fig, axs = plt.subplots(3, 1, figsize=(3, 6))
# 在第一个子图中绘制第一个弦波
axs[0].plot(t, wave_1)
axs[0].set_title('Wave 1')
axs[0].set_ylabel('Amplitude')

# 在第二个子图中绘制第二个弦波
axs[1].plot(t, wave_2)
axs[1].set_title('Wave 2')
axs[1].set_ylabel('Amplitude')

# 在第三个子图中绘制谐波
axs[2].plot(t, harmonic_wave)
axs[2].set_title('Harmonic Wave')
axs[2].set_xlabel('Time (s)')
axs[2].set_ylabel('Amplitude')
plt.tight_layout()
plt.show()
```

# 节拍波

```python
import numpy as np
import matplotlib.pyplot as plt

# 设置节拍波参数
amplitude = 1.0
low_frequency = 20.0
high_frequency = 200.0
duration = 0.1
sample_rate = 1000

# 生成时间序列
t = np.linspace(0, duration, int(duration * sample_rate), endpoint=False)

# 生成节拍波信号
beat_wave = amplitude * np.cos(2 * np.pi * low_frequency * t) * np.cos(2 * np.pi * high_frequency * t)

# 生成包络信号
envelope_1 = np.cos(2 * np.pi * low_frequency * t)
envelope_2 = -np.cos(2 * np.pi * low_frequency * t)

# 绘制节拍波信号和包络信号
plt.plot(t, beat_wave, color='rosybrown', label='Beat Wave')
plt.plot(t, envelope_1, '--', color='y', label='Envelope 1')
plt.plot(t, envelope_2, '--', color='teal', label='Envelope 2')

plt.title('Beat Wave Signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.legend()
plt.axis([0, 0.1, -1, 1])
plt.show()
```

# 淡入与淡出

```python
import numpy as np
import matplotlib.pyplot as plt

# 设置弦波参数
frequency = 5.0
duration = 1.0
sample_rate = 1000

# 生成时间序列
t = np.linspace(0, duration, int(duration * sample_rate), endpoint=False)

# 生成弦波信号
wave = np.cos(2 * np.pi * frequency * t)

# 生成淡入系数数组
fade_in = np.linspace(0, 1, int(duration * sample_rate), endpoint=False)

# 生成淡出系数数组
fade_out = np.linspace(1, 0, int(duration * sample_rate), endpoint=False)

#应用淡入淡出
wave_fade_in = wave * fade_in
wave_fade_out = wave * fade_out

fig, (ax1, ax2) = plt.subplots(1, 2)
plt.subplots_adjust(hspace=0.2, wspace=0.6)
#淡入
ax1.plot(t, wave_fade_in)
ax1.set_title('Fade-In')
ax1.set_xlabel('Time (s)')
ax1.set_ylabel('Amplitude')
ax1.axis([0, 1, -1.2, 1.2])
#淡出
ax2.plot(t, wave_fade_out)
ax2.set_title('Fade-Out')
ax2.set_xlabel('Time (s)')
ax2.set_ylabel('Amplitude')
ax2.axis([0, 1, -1.2, 1.2])
plt.show()
```

# 啁啾信号

```python
import numpy as np
import matplotlib.pyplot as plt

# 生成时间序列
t = np.linspace(0, 3, 2000)

# 线性递增，斜率为50
def frequency_sweep_1(t):
    return 50 * t  

# 正弦波动，中心频率100Hz，振幅100Hz，周期5秒
def frequency_sweep_2(t):
    return 100 * np.sin(2 * np.pi * 0.2 * t) + 100  

# 指数递增，初始频率0Hz，渐近线为200Hz
def frequency_sweep_3(t):
    return 200 * (1 - np.exp(-0.2 * t))  

#根号递增，起始频率10Hz
def frequency_sweep_4(t):
    return 10 + 50 * np.sqrt(t) 

# 生成四种啁啾信号
chirp_signal_1 = np.sin(2 * np.pi * np.cumsum(frequency_sweep_1(t)) / 5000)#线性
chirp_signal_2 = np.sin(2 * np.pi * np.cumsum(frequency_sweep_2(t)) / 5000)#正弦波动
chirp_signal_3 = np.sin(2 * np.pi * np.cumsum(frequency_sweep_3(t)) / 5000)#指数递增
chirp_signal_4 = np.sin(2 * np.pi * np.cumsum(frequency_sweep_4(t)) / 5000)#根号递增

fig, axs = plt.subplots(4, 1, figsize=(8, 10))

#啁啾信号1
axs[0].plot(t, chirp_signal_1)
axs[0].set_title('Chirp Signal 1')
axs[0].set_ylabel('Amplitude')

#啁啾信号2
axs[1].plot(t, chirp_signal_2)
axs[1].set_title('Chirp Signal 2')
axs[1].set_ylabel('Amplitude')

#啁啾信号3
axs[2].plot(t, chirp_signal_3)
axs[2].set_title('Chirp Signal 3')
axs[2].set_ylabel('Amplitude')

#啁啾信号4
axs[3].plot(t, chirp_signal_4)
axs[3].set_title('Chirp Signal 4')
axs[3].set_xlabel('Time (s)')
axs[3].set_ylabel('Amplitude')
plt.tight_layout()
plt.show()
```
