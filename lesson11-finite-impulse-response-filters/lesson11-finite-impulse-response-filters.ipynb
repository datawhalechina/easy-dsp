# 归零滤波器示例

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal as signal

# 生成输入信号（包含原始信号和干扰信号）
fs = 100  # 采样频率
t = np.linspace(0, 1, fs, endpoint=False)  # 时间轴
f1 = 10  # 原始信号频率
f2 = 20  # 干扰信号频率
x = np.cos(2 * np.pi * f1 * t) + np.cos(2 * np.pi * f2 * t)

# 设计归零滤波器（陷波滤波器）
notch_freq = 20  # 要消除的干扰信号频率
b, a = signal.iirnotch(notch_freq, 30, fs)  # 设计陷波滤波器，品质因数Q=30

# 应用滤波器
filtered_x = signal.lfilter(b, a, x)

plt.figure(figsize=(10, 6))

# 原始信号
plt.subplot(2, 1, 1)
plt.plot(t, x)
plt.title('Original Signal')
plt.ylabel('Amplitude')

# 滤波后的信号
plt.subplot(2, 1, 2)
plt.plot(t, filtered_x)
plt.title('Filtered Signal (Nulling Filter)')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.tight_layout()
plt.show()
```
