# IIR滤波器应用

```python
import numpy as np
import matplotlib.pyplot as plt

# IIR滤波器函数
def iir_filter(x):
    y = np.zeros_like(x)
    for n in range(len(x)):
        if n == 0:
            y[n] = 0.8 * y[n] + x[n]
        else:
            y[n] = 0.8 * y[n - 1] + x[n]
    return y

# 生成输入信号
n = np.arange(5)
x = np.array([1, 2, 1, -1, -2])

# 计算IIR滤波器输出
y = iir_filter(x)

plt.figure(figsize=(10, 6))

# 输入信号
plt.subplot(2, 1, 1)
plt.stem(n, x)
plt.title('Input Signal')
plt.ylabel('Amplitude')

# 输出信号
plt.subplot(2, 1, 2)
plt.stem(n, y)
plt.title('Output Signal of IIR Filter')
plt.xlabel('Sample Index')
plt.ylabel('Amplitude')
plt.tight_layout()
plt.show()
```

# IIR滤波器的脉冲响应

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal as signal

# 定义IIR滤波器系数
b = np.array([1])
a = np.array([1, -0.8])

# 生成单位脉冲信号
n = np.arange(10)
impulse = np.zeros_like(n)
impulse[0] = 1

# 计算脉冲响应
impulse_response = signal.lfilter(b, a, impulse)

plt.figure(figsize=(10, 6))

# 输入脉冲信号
plt.subplot(2, 1, 1)
plt.stem(n, impulse)
plt.title('Input Signal (Unit Impulse)')
plt.ylabel('Amplitude')

# 脉冲响应
plt.subplot(2, 1, 2)
plt.stem(n, impulse_response)
plt.title('IIR Filter Impulse Response')
plt.xlabel('Sample Index')
plt.ylabel('Amplitude')

plt.tight_layout()
plt.show()
```

# IIR滤波器的阶跃响应

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal as signal

# 定义IIR滤波器系数
b = np.array([1])
a = np.array([1, -0.8])

# 生成单位阶跃信号
n = np.arange(10)
step = np.ones_like(n)

# 计算阶跃响应
step_response = signal.lfilter(b, a, step)

# 绘制阶跃响应
plt.figure(figsize=(10, 6))

# 原始输入信号（单位阶跃信号）
plt.subplot(2, 1, 1)
plt.stem(n, step)
plt.title('Original Input Signal (Unit Step)')
plt.ylabel('Amplitude')

# 阶跃响应
plt.subplot(2, 1, 2)
plt.plot(n, step_response)
plt.title('IIR Filter Step Response')
plt.xlabel('Sample Index')
plt.ylabel('Amplitude')
plt.stem(n, step_response, use_line_collection=True)
plt.tight_layout()
plt.show()
```
