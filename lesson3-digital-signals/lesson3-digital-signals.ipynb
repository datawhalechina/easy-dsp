# 本章节的代码旨在展示如何将wav音频文件里的音频可视化

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io.wavfile import read
# 读取WAV文件
sample_rate, audio_data = read('xianbo.wav')
num_samples = len(audio_data)
# 生成时间序列
time = np.linspace(0, num_samples / sample_rate, num_samples, endpoint=False)

plt.plot(time, audio_data)
plt.title('Audio Waveform')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.show()
```

# 读取wav文件信息

```python
import wave
import struct
def read_wav_file_info(file_path):
    wav = wave.open(file_path, 'rb')
    num_channels = wav.getnchannels()  # 通道数
    sampwidth = wav.getsampwidth()  # 样本宽度（字节数）
    fs = wav.getframerate()  # 采样频率（Hz）
    num_frames = wav.getnframes()  # 音框数（样本数）
    comptype = wav.getcomptype()  # 压缩类型
    compname = wav.getcompname()  # 压缩名称
    wav.close()

    print(f"通道数: {num_channels}")
    print(f"样本宽度: {sampwidth} 字节")
    print(f"采样频率: {fs} Hz")
    print(f"音框数（样本数）: {num_frames}")
    print(f"压缩类型: {comptype}")
    print(f"压缩名称: {compname}")

    return num_channels, sampwidth, fs, num_frames, comptype, compname
file_path = "digitalsignal.wav"  
num_channels, sampwidth, fs, num_frames, comptype, compname = read_wav_file_info(file_path)
```
